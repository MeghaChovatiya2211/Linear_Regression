```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
from matplotlib import style
style.use('ggplot')
import pandas as pd
```

```python
xs = np.array([1,2,3,4,5] , dtype = np.float64)
ys = np.array([5,4,6,5,6] , dtype = np.float64)
```

```python
def best_fit_slope_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean (xs*ys))/
         ((mean(xs)*mean(xs))-mean(xs*xs)))
    
    b = mean(ys) - m*mean(xs)
    return m,b
    
m,b = best_fit_slope_intercept(xs,ys)

print(m,b)
```

```python
regression_line = []
for x in xs:
    regression_line.append((m*x) + b)
```

```python
plt.scatter(xs,ys)
plt.plot(xs, m*xs + b)
plt.show()
```

```python
def predict(x):
    return (m*x + b)
```

```python
predict(25)
```

```python
df = pd.read_csv('bike sharing hour.csv')
type(df)
```

```python
list(df)
df.drop(['dteday'],1,inplace = True)   #1 means column --> syntax(column_name, axis (0/1), inpace =True/False)
df.drop(['instant'],1,inplace = True)  
full_data = df.astype(float).values.tolist()   # converts all row values to list. Example --> full_data[0] meaning row 0 and all row 0 values will be now stored in a list 

#print(full_data[0])    
```

```python
header = list(df)
```

```python
xs = np.array(full_data)[:,:-1]  # :-1 means all values till column -1 , excluding -1
#print(xs)
ys = np.array(full_data)[:,-1]
#print(ys)
```

```python
m , b = [],[]
for i in range(0,len(xs[0])):
    m1,b1 = best_fit_slope_intercept(xs[:,i],ys)
    m.append(m1)
    b.append(b1)
    plt.scatter(xs[:,i],ys,label = header[i],color = 'g')
    plt.plot(xs[:,i],m1*xs[:,i]+b1,label = header[i])
    plt.legend(loc = 'best')  #loc is used to specify the location of legend. Legend means the signs here on top left. best means top left corner
    plt.savefig('plot_'+header[i]+'.png')  # saves the plot in png image form
    plt.show()
```

```python
print("m =",m)
print("b =",b)
```
